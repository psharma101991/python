Q.1 Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
a=int(input("enter value of a\n"))
if a>0:
    print("number is positive")
elif a==0:
    print("number is 0")
else :
    print("number is negative")
```

Q.4 Write a python program to check prime numbers?

```python
a=int(input("enter value of a\n"))
if a==1:
    print(f"{a} is prime number")
else:
    for i in range(2,a):
        if a%i==0:
            print(f"{a} is not prime number")
            break
    else:
        print(f"{a} is prime number")
```
